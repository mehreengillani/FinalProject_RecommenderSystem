```python
# ============================================
# ANALYZE YOUR DATA SIZE
# ============================================

import json
from collections import Counter
import os

print("="*60)
print("📊 ANALYZING SPORTS & OUTDOORS DATA")
print("="*60)

# File path
file_path = "Sports_and_Outdoors.jsonl"

# Check file size
file_size = os.path.getsize(file_path) / (1024**3)  # Convert to GB
print(f"\n📁 Original file size: {file_size:.2f} GB")

# First pass: Count ratings per user
print("\n📊 Counting ratings per user...")
user_ratings = Counter()
total_ratings = 0

with open(file_path, 'r') as f:
    for i, line in enumerate(f):
        if i % 100000 == 0:
            print(f"  Processed {i:,} lines...")
        try:
            data = json.loads(line.strip())
            user_id = data.get('user_id')
            rating = data.get('rating')
            if user_id and rating is not None:
                user_ratings[user_id] += 1
                total_ratings += 1
        except:
            continue

print(f"\n✅ Total ratings: {total_ratings:,}")
print(f"✅ Total users: {len(user_ratings):,}")

# Count users with 5+ ratings
active_users = sum(1 for count in user_ratings.values() if count >= 5)
print(f"✅ Users with 5+ ratings: {active_users:,} ({active_users/len(user_ratings)*100:.1f}%)")

# How many ratings would we keep?
ratings_kept = sum(count for count in user_ratings.values() if count >= 5)
print(f"✅ Ratings to keep: {ratings_kept:,} ({ratings_kept/total_ratings*100:.1f}%)")
```

```python
# ============================================
# CREATE FILTERED DATASET (5+ REVIEWS)
# ============================================

import json
import os

print("\n" + "="*60)
print("✂️  CREATING FILTERED DATASET")
print("="*60)

# File paths
input_file = "Sports_and_Outdoors.jsonl"
output_file = "Sports_and_Outdoors_filtered_5plus.jsonl"

print(f"\n📥 Input: {input_file}")
print(f"📤 Output: {output_file}")

# Filter and write
print("\n🔄 Filtering data (keeping users with 5+ ratings)...")

lines_written = 0
with open(input_file, 'r') as infile, open(output_file, 'w') as outfile:
    for i, line in enumerate(infile):
        if i % 100000 == 0:
            print(f"  Processed {i:,} lines...")
        try:
            data = json.loads(line.strip())
            user_id = data.get('user_id')
            if user_id and user_ratings.get(user_id, 0) >= 5:
                outfile.write(line)
                lines_written += 1
        except:
            continue

print(f"\n✅ Written {lines_written:,} reviews to filtered file")

# Check new file size
new_size = os.path.getsize(output_file) / (1024**3)  # Convert to GB
print(f"📁 New file size: {new_size:.2f} GB")

# Calculate reduction
reduction = (1 - new_size/file_size) * 100
print(f"📊 Size reduction: {reduction:.1f}%")
```

```python
# ============================================
# CREATE SMALLER SAMPLE 
# ============================================

import random
import json

def create_sample(input_file, output_file, sample_fraction=0.18):
    """
    Create a random sample of the filtered data
    """
    print(f"\n📊 Creating {sample_fraction*100:.0f}% sample...")
    
    lines_written = 0
    with open(input_file, 'r') as infile, open(output_file, 'w') as outfile:
        for line in infile:
            if random.random() < sample_fraction:
                outfile.write(line)
                lines_written += 1
    
    print(f"✅ Written {lines_written:,} reviews")
    
    # Check new size
    size = os.path.getsize(output_file) / (1024**3)
    print(f"📁 Sample file size: {size:.2f} GB")
    
    return size

# Create a 18% sample
sample_file = "Sports_and_Outdoors_sample_18pct.jsonl"
sample_size = create_sample(output_file, sample_file, 0.18)
```

```python
# ============================================
# FILTER METADATA FOR SAMPLE PRODUCTS
# ============================================

print("="*60)
print("📊 FILTERING METADATA FOR SAMPLE")
print("="*60)

# Get product IDs from reviews sample
print("\n🔍 Extracting product IDs from reviews...")
product_ids = set()

with open("Sports_and_Outdoors_sample_18pct.jsonl", 'r') as f:
    for line in f:
        try:
            data = json.loads(line.strip())
            asin = data.get('parent_asin')
            if asin:
                product_ids.add(asin)
        except:
            continue

print(f"✅ Found {len(product_ids):,} unique products in review sample")

# Filter metadata
print("\n📚 Filtering metadata...")
metadata_file = "meta_Sports_and_Outdoors.jsonl"
filtered_meta = "meta_Sports_and_Outdoors_sample.jsonl"

lines_written = 0
with open(metadata_file, 'r') as infile, open(filtered_meta, 'w') as outfile:
    for line in infile:
        try:
            data = json.loads(line.strip())
            asin = data.get('parent_asin')
            if asin in product_ids:
                outfile.write(line)
                lines_written += 1
        except:
            continue

print(f"✅ Written {lines_written:,} metadata records")

# Check size
size_mb = os.path.getsize(filtered_meta) / (1024**2)
print(f"📁 Metadata size: {size_mb:.1f} MB")
```

```python
# ============================================
# DATA QUALITY CHECK
# ============================================

print("\n" + "="*60)
print("📊 DATA QUALITY CHECK")
print("="*60)

# Check reviews
print("\n📊 Reviews Sample:")
reviews_stats = {'ratings': [], 'users': set(), 'products': set()}

with open("Sports_and_Outdoors_sample_18pct.jsonl", 'r') as f:
    for i, line in enumerate(f):
        if i >= 50000:  # Check first 50K
            break
        try:
            data = json.loads(line.strip())
            reviews_stats['ratings'].append(data.get('rating', 0))
            reviews_stats['users'].add(data.get('user_id'))
            reviews_stats['products'].add(data.get('parent_asin'))
        except:
            continue

print(f"  Avg Rating: {sum(reviews_stats['ratings'])/len(reviews_stats['ratings']):.2f}")
print(f"  Users: {len(reviews_stats['users']):,}")
print(f"  Products: {len(reviews_stats['products']):,}")

# Check metadata
print("\n📚 Metadata Sample:")
meta_stats = {'titles': 0, 'descriptions': 0, 'prices': 0, 'categories': 0}

with open(filtered_meta, 'r') as f:
    for i, line in enumerate(f):
        if i >= 50000:
            break
        try:
            data = json.loads(line.strip())
            if data.get('title'): meta_stats['titles'] += 1
            if data.get('description'): meta_stats['descriptions'] += 1
            if data.get('price'): meta_stats['prices'] += 1
            if data.get('categories'): meta_stats['categories'] += 1
        except:
            continue

total = min(50000, lines_written)
print(f"  Titles: {meta_stats['titles']/total*100:.1f}%")
print(f"  Descriptions: {meta_stats['descriptions']/total*100:.1f}%")
print(f"  Prices: {meta_stats['prices']/total*100:.1f}%")
print(f"  Categories: {meta_stats['categories']/total*100:.1f}%")
```

```python
# ============================================
# FINAL VERIFICATION
# ============================================

print("\n" + "="*60)
print("✅ PROJECT READY CHECK")
print("="*60)

files = [
    ("Reviews", "Sports_and_Outdoors_sample_18pct.jsonl"),
    ("Metadata", "meta_Sports_and_Outdoors_sample.jsonl")
]

for name, path in files:
    if os.path.exists(path):
        size_mb = os.path.getsize(path) / (1024**2)
        print(f"✅ {name}: {size_mb:.1f} MB")
    else:
        print(f"❌ {name}: Not found")

print("\n" + "="*60)
print("📊 DATASET SUMMARY")
print("="*60)

# Count lines in reviews
with open("Sports_and_Outdoors_sample_18pct.jsonl", 'r') as f:
    review_count = sum(1 for _ in f)
print(f"Total Reviews: {review_count:,}")

# Count lines in metadata
with open("meta_Sports_and_Outdoors_sample.jsonl", 'r') as f:
    meta_count = sum(1 for _ in f)
print(f"Metadata Records: {meta_count:,}")

print(f"\n✅ Dataset is ready for modeling!")
print(f"   - Reviews: {review_count:,} (~480MB)")
print(f"   - Products with metadata: {meta_count:,}")
print(f"   - Good for: Databricks, Colab, or local Spark")
```

```python
# ============================================
# COMPLETE DATASET PREPARATION - AGGGRESSIVE
# ============================================

import json
import os
from collections import Counter

print("="*60)
print("📊 SPORTS & OUTDOORS - DATASET PREPARATION")
print("="*60)

# ============================================
# 1. CREATE FILTERED REVIEWS (USERS WITH 5+ RATINGS)
# ============================================

print("\n📊 Step 1: Filtering users with 5+ ratings...")

input_reviews = "Sports_and_Outdoors.jsonl"
output_reviews = "Sports_and_Outdoors_filtered.jsonl"

# Count ratings per user
user_ratings = Counter()
with open(input_reviews, 'r') as f:
    for line in f:
        try:
            data = json.loads(line.strip())
            user = data.get('user_id')
            if user:
                user_ratings[user] += 1
        except:
            continue

active_users = {u for u, c in user_ratings.items() if c >= 5}
print(f"✅ Active users: {len(active_users):,}")

# Write filtered reviews
count = 0
with open(input_reviews, 'r') as infile, open(output_reviews, 'w') as outfile:
    for line in infile:
        try:
            data = json.loads(line.strip())
            if data.get('user_id') in active_users:
                outfile.write(line)
                count += 1
        except:
            continue

print(f"✅ Reviews kept: {count:,}")

# ============================================
# 2. CREATE 18% SAMPLE
# ============================================

print("\n📊 Step 2: Creating 18% sample...")

import random
random.seed(42)

sample_file = "Sports_and_Outdoors_sample.jsonl"
sample_count = 0

with open(output_reviews, 'r') as infile, open(sample_file, 'w') as outfile:
    for line in infile:
        if random.random() < 0.18:
            outfile.write(line)
            sample_count += 1

print(f"✅ Sample reviews: {sample_count:,}")

# ============================================
# 3. FILTER METADATA - ONLY ESSENTIAL FIELDS
# ============================================

print("\n📊 Step 3: Filtering metadata (essential fields only)...")

# Get product IDs from sample
product_ids = set()
with open(sample_file, 'r') as f:
    for line in f:
        try:
            data = json.loads(line.strip())
            asin = data.get('parent_asin')
            if asin:
                product_ids.add(asin)
        except:
            continue

print(f"✅ Unique products: {len(product_ids):,}")

# Filter metadata - keep only essential fields
input_meta = "meta_Sports_and_Outdoors.jsonl"
output_meta = "meta_filtered.jsonl"

meta_count = 0
with open(input_meta, 'r') as infile, open(output_meta, 'w') as outfile:
    for line in infile:
        try:
            data = json.loads(line.strip())
            asin = data.get('parent_asin')
            if asin in product_ids:
                # Keep only essential fields
                filtered_data = {
                    'parent_asin': asin,
                    'title': data.get('title', ''),
                    'description': data.get('description', '')[:500],  # Truncate description
                    'categories': data.get('categories', [])[:3],      # First 3 categories only
                    'price': data.get('price', None),
                    'average_rating': data.get('average_rating', None)
                }
                outfile.write(json.dumps(filtered_data) + '\n')
                meta_count += 1
        except:
            continue

print(f"✅ Metadata records: {meta_count:,}")

# ============================================
# 4. FINAL SUMMARY
# ============================================

print("\n" + "="*60)
print("📊 FINAL SUMMARY")
print("="*60)

files = [
    ("Reviews Sample", "Sports_and_Outdoors_sample.jsonl"),
    ("Filtered Metadata", "meta_filtered.jsonl")
]

total_size = 0
for name, path in files:
    if os.path.exists(path):
        size_mb = os.path.getsize(path) / (1024**2)
        total_size += size_mb
        line_count = sum(1 for _ in open(path, 'r'))
        print(f"\n{name}:")
        print(f"  Records: {line_count:,}")
        print(f"  Size: {size_mb:.1f} MB")

print(f"\n📁 Total upload size: {total_size:.1f} MB")

if total_size < 500:
    print("✅ Under 500MB - Ready for Databricks!")
else:
    print(f"⚠️ Over 500MB by {total_size - 500:.1f} MB")
    print("\n📊 If still too large, create smaller sample:")
    print("   - Create 15% sample instead of 18%")
    print("   - Or compress to GZIP before uploading")

print("\n" + "="*60)
print("✅ DATASET READY")
print("="*60)
print("\n📤 Upload these two files:")
print("  1. Sports_and_Outdoors_sample.jsonl")
print("  2. meta_filtered.jsonl")
```
